##**TEAM DISTRIBUTION AND PERCENTAGE SPLITTING RELATIVE TO TOTAL EMPLOYEES**

```python
plt.figure(figsize=(30,7))
plt.xticks(rotation=90)
sns.countplot(x='Team',data=df_employee,hue= 'Team',palette='mako')
plt.show()
```

```python
# Count the number of employees in each team
team_counts = df_employee["Team"].value_counts()

# Calculate the total number of employees
total_employees = len(df_employee)

# Calculate the percentage of employees in each team
percentage = (team_counts / total_employees) * 100

# Create a DataFrame to display results
result = pd.DataFrame({'No of Employees': team_counts, 'Percentage': percentage.round(2)})

# Display the result
print(result)
```

##**Find out under which team and position, spending in terms of salary is high.**

```python
df_employee["Salary"].value_counts()
```

##**Find if there is any correlation between age and salary , represent it visually**
###correlation:

*Values close to +1 indicates strong positive correlation, -1 indicates a strong negative correlation and 0 indicates suggests no linear correlation. Darker colors signify strong correlation, while light colors represents weaker correlations. Positive correlation variable move in same directions. As one increases, the other also increases. Negative correlation variable move in opposite directions. An increase in one variable is associated with a decrease in the other.*

###heat Maps

*Heat Maps is a type of plot which is necessary when we need to find the dependent variables. One of the best way to find the relationship between the features can be done using heat mapsIn*

```python
df2 = df_employee[["Age","Salary"]]
df2
```

```python
corr = df2.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="mako")
plt.title("correlation matrix between age and salary of employees")
plt.show()
```

##**Data Story:**

1. *Age vs. Age (1.0000): The value of 1.0000 along the diagonal signifies a perfect positive correlation, as expected.*

2. *Age vs. Salary (0.21117): The correlation coefficient of 0.21117 indicates a positive but weak relationship between age and salary. With increasing age, there's a slight tendency for salaries to rise, but the correlation is not very strong.*

3. *Salary vs. Age (0.21117): This symmetric measure mirrors the correlation between "Age" and "Salary," reaffirming the weak positive relationship between the two variables.*

*In essence, the correlation matrix reveals a positive yet relatively weak association between age and salary in the employee dataset. As employees' ages increase, there's a tendency for their salaries to slightly rise.8
